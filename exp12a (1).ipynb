```python
import pandas as pd

df = pd.read_csv("/content/Expt11.csv")

display(df)
```

```python
#12.1.1 Number of Students in each Department, Each Gender and Each Grade


print(df['Grade'].value_counts())

print(df['Gender'].value_counts())

print(df['Department'].value_counts())
```

```python
#12.1.2 Percentage distribution of Grades

display("Grade Percentage")
display(df['Grade'].value_counts(normalize=True)*100)
```

```python
#12.1.3 Gender vs Grade Analysis, Department vs Gender Analysis, Department vs Grade Analysis

display(pd.crosstab(df['Gender'],df['Grade']))

display(pd.crosstab(df['Department'],df['Gender']))

display(pd.crosstab(df['Department'],df['Grade']))
```

```python
#12.1.4 Department vs Grade Percentage Analysis

display(pd.crosstab(df['Department'], df['Grade'], normalize='index')*100)
```

```python
#12.1.5 Grouping Data

display(df.groupby('Department')['Grade'].value_counts())
```

```python
import pandas as pd

data = {
'Order_ID':['01','02','03','04','05','06','07','08','09','10'],

'Category':['Electronics','Clothing','Electronics','Grocery','Clothing',
            'Electronics','Grocery','Clothing','Electronics','Grocery'],

'Payment_Method':['UPI','Card','UPI','Cash','Card',
                  'UPI','Cash','Card','UPI','Cash'],

'Delivery_Type':['Express','Standard','Express','Standard','Express',
                 'Standard','Express','Standard','Express','Standard'],

'Customer_Type':['New','Returning','Returning','New','Returning',
                 'New','Returning','New','Returning','New']
}

df = pd.DataFrame(data)

display(df)
```

```python
#12.2.1 Frequency Count

display(df['Category'].value_counts())
```

```python
#12.2.2 Percentage Distribution

display(df['Category'].value_counts(normalize=True)*100)
```

```python
#12.2.3 Payment Method Analysis

display(df['Payment_Method'].value_counts())
```

```python
#12.2.4 Unique Values

print(df['Category'].unique())
```

```python
#12.2.5 no. of unique Values

print(df['Category'].nunique())
```

```python
#12.2.6 Filtering Data

electronics_orders = df[df['Category']=='Electronics']

display(electronics_orders)
```

```python
#12.2.7 Sorting Data

sorted_df = df.sort_values(by='Category')
display(sorted_df)
```

```python
#12.2.8 Grouping

display(df.groupby('Category')['Payment_Method'].value_counts())
```
